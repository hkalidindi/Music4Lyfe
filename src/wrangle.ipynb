### Wrangle

Used exclusively to clean Spotify data from ```src/resources/spotify_playlist.csv```. Output will be stored in ```src/resources/spotify_cleaned.csv```.

````python
import pandas as pd
````

````python
df = pd.read_csv('./resources/spotify_playlist.csv')
df
````

````python
def clean_col(s: str) -> str:
    return "_".join(s.strip().split(' ')).lower()

cols = df.columns
cleaned_cols = list(map(clean_col, cols))
rename = dict(zip(cols, cleaned_cols))
df = df.rename(columns = rename)

want_to_drop = ["playlist_name", "type", "isrc"]

df = df.drop(columns=want_to_drop, axis=1)
df
````

````python
df.to_csv('./resources/spotify_cleaned.csv', index=False)
````
